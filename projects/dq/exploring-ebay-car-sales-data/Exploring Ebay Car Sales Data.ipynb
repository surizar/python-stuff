```python
#removing $ symbol and ','.  Then convert to float
autos["price"] = autos["price"].str.replace("$","")
autos["price"] = autos["price"].str.replace(",","")
#convert to float
autos["price"] = autos["price"].astype(float)
```

```python
# odometer remove km and ','. Then convert to int and rename column
autos["odometer"] = autos["odometer"].str.replace("kms","")
autos["odometer"] = autos["odometer"].str.replace("km","")
autos["odometer"] = autos["odometer"].str.replace(",","")

#conver to int
autos["odometer"] = autos["odometer"].astype(int)

#rename column to odometer_km
autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)
```

According to the previous data, there is a min registration year of 1000 and a max of 9999 which are out of the valid range. 
Year 1k was long time ago and I'd need a time machine to check if cars exist in 9999 :-)

Assuming a minimum year of 1900 and a max of 2016 (dataset year), the valid car range would be 1900 to 2016.

Below, I'm checking how many cars are outside of the valid registration year range.

```python
bool_reg_year = autos["registration_year"].between(1900,2016)

autos_in_range = autos.loc[bool_reg_year,:]

top_15=autos_in_range["registration_year"].value_counts(normalize=True).sort_values(ascending=False)[:15]
print(top_15)
#normalize = True
#With normalize set to True, 
#returns the relative frequency by dividing all values 
#by the sum of values.
#ref: https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html
print(sum(top_15))
```

From the data above, 77.9% of the vehicles were registered from 1997 to 2011

#### Exploring price by brand


```python
autos["brand"].value_counts(normalize=True).sort_values(ascending=False)
```

```python
top_10_brands = autos["brand"].value_counts(normalize=True).sort_values(ascending=False)[:10]
print(top_10_brands)
print(sum(top_10_brands))
```

80% of the vehicles are in the top 10 brands of the list.
I'm focusing on those 10 first to get the average price by brand

```python
price_mean_by_brand = {}

for brand in top_10_brands.index:
    current_brand= autos["brand"]== brand
    mean = autos.loc[current_brand,"price"].mean()
    price_mean_by_brand[brand] = int(mean)
    
price_mean_by_brand
```

```python
dict(sorted(price_mean_by_brand.items(), key=lambda item: item[1]))
```

* Audi, MB and Audi are above 8k, those are the more expensive cars in avg.
* While VW at 5.3k avg price (VW has the largest share ~21.31%)
* renault, fiat and opel are among the less expensive cars ( < 3k )
* Ford is one of the non european top 6 cars, near to the less expensive cars( 3.7k ) and ranked in 6th place on the brand popularity rank.

#### Exploring car mileage


```python
#Calculating mean mileage
mileage_mean_by_brand = {}

for brand in top_10_brands.index:
    current_brand = autos["brand"]== brand
    mean = autos.loc[current_brand,"odometer_km"].mean()
    mileage_mean_by_brand[brand] = int(mean)
    
mileage_mean_by_brand
```

```python
dict(sorted(mileage_mean_by_brand.items(), key=lambda item: item[1]))
```

```python
bpm_mean = pd.Series(price_mean_by_brand)
bpm_mean = bpm_mean.sort_values(ascending=False)
bpm_df = pd.DataFrame(bpm_mean, columns=['mean_price'])
bpm_df
```

```python
mmb_mean = pd.Series(mileage_mean_by_brand)
mmb_mean = mmb_mean.sort_values(ascending=False)
mmb_df = pd.DataFrame(mmb_mean, columns=['mean_odometer_km'])
mmb_df
```

```python
results_df = bpm_df
results_df["mean_odometer"]=mmb_mean
results_df
```

When comparing prices and avg kms(odometer), cars with higher prices are almost also the cars with more kms.  This don't necesary means that those cars lasts more, it could also be an indicator that luxury car owners tend to use their cars for more time.
