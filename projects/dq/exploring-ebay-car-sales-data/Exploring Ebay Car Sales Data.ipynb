```python
#removing $ symbol and ','.  Then convert to float
autos["price"] = autos["price"].str.replace("$","")
autos["price"] = autos["price"].str.replace(",","")
#convert to float
autos["price"] = autos["price"].astype(float)
```

```python
# odometer remove km and ','. Then convert to int and rename column
autos["odometer"] = autos["odometer"].str.replace("kms","")
autos["odometer"] = autos["odometer"].str.replace("km","")
autos["odometer"] = autos["odometer"].str.replace(",","")

#conver to int
autos["odometer"] = autos["odometer"].astype(int)

#rename column to odometer_km
autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)
```

```python
autos.head()
```

#### Analyzing odometer and price

```python
odometer = autos["odometer_km"]

print (min(odometer))
print (max(odometer))
print (odometer.unique().shape)
print (odometer.describe())
print (odometer.value_counts().sort_index(ascending=False))
```

```python
price = autos["price"]

print (min(price))
print (max(price))
print (price.unique().shape)
print (price.describe())
print (price.value_counts().sort_index(ascending=False))
print (price.value_counts().sort_index(ascending=False).head(10))
```

```python
autos_bool_price = autos["price"].between(100,100000)
filtered_autos = autos[autos_bool_price]
print(filtered_autos["price"].describe())
```

#### Removing outliers
I'm removing cars with prices above 100k to avoid outliers.  The removed cars represents around 3.6% of the total data.

```python
autos = filtered_autos
print(autos["price"].describe())
```

```python
#converting to date and finding % distribution
date_crawled = autos['date_crawled']
date_crawled = date_crawled.str[:10]
data_crawled = date_crawled.value_counts(normalize=True, dropna=False)
#print (100*data_crawled/date_crawled.value_counts().sum())
print (data_crawled)
```

```python
ad_created = autos['ad_created']
ad_created = ad_created.str[:10]
ad_created = ad_created.value_counts()
print (ad_created)
print (ad_created/ad_created.value_counts().sum())

last_seen = autos['last_seen']
last_seen = last_seen.str[:10]
last_seen = last_seen.value_counts()
print (last_seen)
print (last_seen.value_counts().sum())
print (last_seen/last_seen.value_counts().sum())
```
